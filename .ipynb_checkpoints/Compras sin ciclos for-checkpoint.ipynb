```python
%reset -f 

import numpy as np
import tkinter as tk
from tkinter import ttk
import datetime as dt
from datetime import datetime
from datetime import timedelta
import json
import jsonpickle
lista_recetas=["Arroz frito","Milanesa de pechuga","Puré de papas","<--->"]

def listas_compras():
    pass

def my_command():
    pass
    
def abre_cronograma_default():
    cronograma_default_window=tk.Tk()
    WIDTH = 800
    HEIGHT = 500
    x = int((cronograma_default_window.winfo_screenwidth() / 2) - (WIDTH / 2))
    y = int((cronograma_default_window.winfo_screenheight() / 2) - (HEIGHT /2 ))
    cronograma_default_window.geometry(f'{WIDTH}x{HEIGHT}+{x}+{y}')
    cronograma_default_window.title('Cronograma default')
    
    menubar_cronograma_default=tk.Menu(cronograma_default_window)
    cronograma_default_window.config(menu=menubar_cronograma_default)
    menubar_cronograma_default.add_command(label='Guardar',command=my_command)
    menubar_cronograma_default.add_command(label='Configurar',command=abre_configurar)
    menubar_cronograma_default.add_command(label='Transferir',command=my_command)
    menubar_cronograma_default.add_command(label='Crear lista',command=my_command)

    variable1_cd=tk.StringVar()
    variable2_cd=tk.StringVar()
    
    ttk.Style(cronograma_default_window).configure('TLabel',background="dark orange")
    celda_1cd= ttk.OptionMenu(cronograma_default_window,variable1_cd,'',style='TLabel',*lista_recetas,command=my_command)
    celda_1cd.place(x=0,y=25,width=100,height=25)
    celda_2cd= ttk.OptionMenu(cronograma_default_window,variable2_cd,'',style='TLabel',*lista_recetas,command=my_command)
    celda_2cd.place(x=0,y=50,width=100,height=25)      
    
    fecha_inicial_cu=dt.datetime(2022, 7, 15)
    fecha_inicial=fecha_inicial_cu.strftime("%d/%m/%Y")
    tk.Label(cronograma_default_window,bg= "black", fg= "white",text=fecha_inicial).place(x=0,y=0,width=100,height=25)
    today = datetime.now()
    today = today.strftime("%d/%m/%Y")
    xx=100
    for i in np.arange(7):
        i=int(i)
        tiempo = fecha_inicial_cu+dt.timedelta(days = i+1)
        tiempo = tiempo.strftime("%d/%m/%Y")
        if (tiempo==today):
            label=tk.Label(cronograma_default_window,bg= "green", fg= "white",text=today).place(x=xx,y=0,width=100,height=25)           
        else:
            label=tk.Label(cronograma_default_window,bg= "black", fg= "white",text=tiempo).place(x=xx,y=0,width=100,height=25)
        xx=xx+100        


        

        
def abre_configurar():
    configurar_window=tk.Tk()
       
    WIDTH = 200
    HEIGHT = 100
    x = int((configurar_window.winfo_screenwidth() / 2) - (WIDTH / 2))
    y = int((configurar_window.winfo_screenheight() / 2) - (HEIGHT /2 ))
    configurar_window.geometry(f'{WIDTH}x{HEIGHT}+{x}+{y}')
    
    configurar_window.title('')
    
    tk.Label(configurar_window,bg= "black", fg= "white",text='Días').place(x=0,y=25,width=100,height=25)
    tk.Label(configurar_window,bg= "black", fg= "white",text='Fecha fin').place(x=0,y=50,width=100,height=25)
    
    entry_configurar(configurar_window)
    


def entry_configurar(configurar_window): 
    tk.Label(configurar_window,bg= "black", fg= "white",text='Fecha inicio').place(x=0,y=0,width=100,height=25)
    
    global entry_fechainicio
    global entry_dias
    global entry_fechafin
    
    entry_fechainicio=tk.Entry(configurar_window) 
    entry_fechainicio.place(x=100,y=0,width=100,height=25)
    
    entry_dias=tk.Entry(configurar_window)   
    entry_dias.place(x=100,y=25,width=100,height=25)  
    
    entry_fechafin=tk.Entry(configurar_window)   
    entry_fechafin.place(x=100,y=50,width=100,height=25)  
    
    boton=tk.Button(configurar_window,text='Aplicar',command=get_fecha)
    boton.place(x=0,y=75,width=200,height=25)  


def get_fecha():
    fechainicio=entry_fechainicio.get()
    dias=entry_dias.get()
    fechafin=entry_fechafin.get()
    
    print(fechainicio)
    print(dias)
    print(fechafin)

    
    
def display_selected1(variable_seleccionada):
    global variable1m
    variable1m=variable_seleccionada
    
def display_selected2(variable_seleccionada):
    global variable2m
    variable2m=variable_seleccionada
    
def guardar_cronograma():
        with open("variables.json", 'w') as outfile_var:
          
            variable_1m_encode=jsonpickle.encode(variable1m)
            variable_2m_encode=jsonpickle.encode(variable2m)            
            
            variables_encode=[variable_1m_encode,variable_2m_encode]
            
            json.dump(variables_encode, outfile_var)    
                         
def cargar_cronograma():
        with open('variables.json', 'r') as openfile_var:
            variables_decode = json.load(openfile_var)
            
            variable_1m_decode=jsonpickle.decode(variables_decode[0])
            variable_2m_decode=jsonpickle.decode(variables_decode[1])         
        
        print(variables_decode[0])
        print(variables_decode[1])
        
        variable1=tk.StringVar()
        variable2=tk.StringVar()
        
        celda_1m= ttk.OptionMenu(main_window,variable1,variable_1m_decode,style='TLabel',*lista_recetas,command=my_command)
        celda_1m.place(x=0,y=25,width=100,height=25)
        celda_2m= ttk.OptionMenu(main_window,variable2,variable_2m_decode,style='TLabel',*lista_recetas,command=my_command)
        celda_2m.place(x=0,y=50,width=100,height=25)  
            


def abre_recetas():
    recetas_window=tk.Tk()
    recetas_window.geometry("400x400")
    recetas_window.title('Recetas')

def abre_reservas():
    reservas_window=tk.Tk()
    reservas_window.geometry("400x400")
    reservas_window.title('Reservas')


    
def labels_cu(main_window):
    fecha_inicial_cu=dt.datetime(2022, 7, 15)
    fecha_inicial=fecha_inicial_cu.strftime("%d/%m/%Y")
    tk.Label(main_window,bg= "black", fg= "white",text=fecha_inicial).place(x=0,y=0,width=100,height=25)
    today = datetime.now()
    today = today.strftime("%d/%m/%Y")
    xx=100
    for i in np.arange(7):
        i=int(i)
        tiempo = fecha_inicial_cu+dt.timedelta(days = i+1)
        tiempo = tiempo.strftime("%d/%m/%Y")
        if (tiempo==today):
            label=tk.Label(main_window,bg= "green", fg= "white",text=today).place(x=xx,y=0,width=100,height=25)           
        else:
            label=tk.Label(main_window,bg= "black", fg= "white",text=tiempo).place(x=xx,y=0,width=100,height=25)
        xx=xx+100    

def create_main_window():
    main_window=tk.Tk()
    main_window.title("Cronograma") 
    main_window.resizable(False, False) 
    
    WIDTH = 800
    HEIGHT = 500
    x = int((main_window.winfo_screenwidth() / 2) - (WIDTH / 2))
    y = int((main_window.winfo_screenheight() / 2) - (HEIGHT /2 ))
    main_window.geometry(f'{WIDTH}x{HEIGHT}+{x}+{y}')

    
    menubar=tk.Menu(main_window) # Crea el menu llamado menubar en main_window
    main_window.config(menu=menubar) # Vincula el menu menubar con main_window
           
    menubar.add_command(label='Recetas',command=abre_recetas)
    menubar.add_command(label='Reservas',command=abre_reservas)
    menubar.add_command(label='Cronograma default',command=abre_cronograma_default)
    menubar.add_command(label='Guardar',command=guardar_cronograma)
    menubar.add_command(label='Cargar',command=cargar_cronograma)
    menubar.add_command(label='Listas',command=listas_compras)
   
    
   
    
    labels_cu(main_window)
    celdas_cu(main_window)
    botones_reservar(main_window)
    return main_window

def celdas_cu(main_window):

    # Cronograma. Primera celda
    x_c=0
    y_c=25
    variable1=tk.StringVar()
    celda1= ttk.OptionMenu(main_window,variable1,'',*lista_recetas,style='TLabel',command=display_selected1)
    celda1.place(x=x_c,y=y_c,width=100,height=25)
    ttk.Style(main_window).configure('TLabel',background="dark orange")
    #print(celda1)

    # Cronograma. Segunda celda
    x_c=0
    y_c=50
    variable2=tk.StringVar()
    celda2= ttk.OptionMenu(main_window,variable2,'',*lista_recetas,style='TLabel',command=display_selected2)
    celda2.place(x=x_c,y=y_c,width=100,height=25)
     
def botones_reservar(main_window):    
    boton1=tk.Button(main_window,text='Reservar',command=my_command)
    boton1.place(x=100,y=25,width=50,height=25)

    boton2=tk.Button(main_window,text='Reservar',command=my_command)
    boton2.place(x=100,y=50,width=50,height=25)

main_window=create_main_window()
main_window.mainloop()
```
